## collaborative filtering model using KNNWithMeans item_based

```python
from surprise import KNNWithMeans
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import train_test_split
from surprise import Reader

df_co = Dataset.load_from_df(df_fifty[['author','product','score']], Reader(rating_scale=(1, 10)))
trainset, testset = train_test_split(df_co, test_size=.20,random_state=612)
```

```python
algo = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': False})
algo.fit(trainset)
```

```python
test_pred = algo.test(testset)
```

```python
# get RMSE
print("Item-based Model : Test Set")
accuracy.rmse(test_pred, verbose=True)
```

## collaborative filtering model using KNNWithMeans user_based

```python
algo1 = KNNWithMeans(k=50, sim_options={'name': 'pearson_baseline', 'user_based': True})
algo1.fit(trainset)
```

```python
test_pred1= algo1.test(testset)
```

```python
# get RMSE
print("Item-based Model : Test Set")
accuracy.rmse(test_pred1, verbose=True)
```

## a collaborative filtering model using SVD

```python
from surprise import SVD
from surprise.model_selection import cross_validate

svd = SVD(random_state=612)
svd.fit(trainset)
svd_pred = svd.test(testset)
```

```python
accuracy.rmse(svd_pred)
```
